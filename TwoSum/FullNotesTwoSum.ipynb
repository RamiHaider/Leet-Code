# TwoSum Problem in LeetCode

The Problem:
Given a list of integers:
iterable = [2,7,11,15,1,9,4,29]

Checking every pair in order in the list
r = 2

We want to find where the target sum of those two is
target = 33

Whilst achieving a time complexity of less than 0(n^2)

I.e., output:
"The goal is to output Indices: (6,7), Values: (4,29)"
In a time less than O(n^2)


Firstly, the core of this problem via the brute force method is to
sum the unique, non repeatable 2-pair combinations in a list

i.e., for:
[a,b,c,d,e]

we need,

[a,b], [a,c], [a,d], [a,e]
[b,c], [b,d], [b,e],
[c,d], [c,e]
[d,e]



```python
def find_two_pair_combinations(input_list, target):
    n = len(input_list)

    for i in range(n-1):
        for j in range(i+1, n):

            if input_list[i]+input_list[j] == target:
                result = [i,j]

    print(result)

my_list = [4, 3, 5, 1, 9, 10, 12, 15]
my_target = 10
find_two_pair_combinations(my_list, my_target)
```

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:

        n = len(nums)

        #length constraints
        if n < 2 or n > 10**4:  
            print("Length of the list is out of bounds")
            return [] 
        
        #target constraints
        if target < -10**9 or target > 10**9:
            print("Target is out of bounds")
            return []
        

        for i in range(n-1):
            for j in range(i+1, n):

                #max value in list contraints
                if nums[i] < -10**9 or nums[i] > 10**9:
                    print(f"The value of {nums[i]} is out of bounds")
                    return []
                if nums[j] < -10**9 or nums[j] > 10**9:
                    print(f"The value of {nums[j]} is out of bounds")
                    return []                  

                if nums[i]+nums[j] == target:
                    return [i,j]
            
        print(f'No target value found within the list')
        return []
```

```python
test = Solution()
print(test.twoSum([2,7,11,15],9))
print(test.twoSum([2],5))
print(test.twoSum([2,5],6))
print(test.twoSum([100000000000000,4,5,5],3))
```

# Nested Loop Approach
Let's analyze the nested loop approach for the list [2, 7, 11, 15] with target 26:

for i in range(n-1):
    for j in range(i+1, n):
        if nums[i] + nums[j] == target:
            return [i, j]
            
Operations per iteration:

Array indexing (nums[i], nums[j]): 2 operations
Addition (nums[i] + nums[j]): 1 operation
Comparison (== target): 1 operation

Total operations per inner loop iteration: 4 

Number of iterations:
1st iteration: 3 comparisons
2nd iteration: 2 comparisons
3rd iteration: 1 comparison

Total iterations: 3 + 2 + 1 = 6
Total operations: 6 * 4 = 24
Worst case (target not found): 24 operations


Rather, what we need is a hashing approach, which will reduce our complexity, by an exponent 2 factorfold.

Step-by-step walkthrough:


Start with an empty dictionary: num_dict = {}
Our list is: [2,7,11,15]
Our target is: 26

### First iteration (i = 0, num = 2):

1. Step 1) Calculate complement: 26 - 2 = 24
2. Step 2) Check:                Is 24 in num_dict? Since No, then:
3. Step 3) Add 2 to num_dict: 

num_dict = {2: 0}


### Second iteration (i = 1, num = 7):

1. Step 1) Calculate complement: 26 - 7 = 19
2. Step 2) Check:                Is 19 in num_dict? Since No, then:
3. Step 3) Add 7 to num_dict: 

num_dict = {2:0, 7: 1}


### Third iteration (i = 2, num = 11):

1. Step 1) Calculate complement: 26 - 11 = 15
2. Step 2) Check:                Is 15 in num_dict? Since No, then:
3. Step 3) Add 11 to num_dict: 

num_dict = {2:0, 7: 1, 11: 2}


### Fourth iteration (i = 3, num = 15):

1. Step 1) Calculate complement: 26 - 15 = 11
2. Step 2) Check:                Is 11 in num_dict? Since Yes
3. Step 3)  

num_dict = {2:0, 7: 1, 11: 2}


Fourth iteration (i = 3, num = 15):

Calculate complement: 26 - 15 = 11
Is 11 in num_dict? Yes! We found a match.
Return [num_dict[11], i] which is [2, 3]




# Time Complexity of the Hash Table Approach
Now let's analyze the hash table approach:
    
Operations per iteration:
Subtraction (target - num): 1 operation
Hash table lookup (complement in num_dict): 1 operation (average case)
Hash table insertion (num_dict[num] = i): 1 operation (average case)

Total operations per iteration: 3
Number of iterations: 4 (we go through the entire list once)
Total operations: 4 * 3 = 12
Worst case (target not found): 12 operations


Memory Storage and Access: Lists vs Dictionaries
Lists in Memory

Contiguous Memory: Lists in Python (and arrays in many languages) are stored in contiguous blocks of memory.
Index-Based Access: Each element can be accessed directly if you know its index, using pointer arithmetic.
Search Process: When searching for a value (not an index), the computer must check each element sequentially.

Dictionaries (Hash Tables) in Memory

Array of Buckets: A dictionary is typically implemented as an array of "buckets".
Hash Function: When you add a key-value pair, the key is passed through a hash function.
Bucket Assignment: The hash function's output determines which bucket the pair goes into.
Collision Handling: If multiple keys hash to the same bucket, the bucket may contain a linked list or another structure to handle collisions.

How "Direct Access" Works in Dictionaries

Hash Computation: When you look up a key, its hash is computed.
Bucket Location: This hash directly corresponds to a bucket's location in memory.
Quick Retrieval: The computer can jump straight to this memory location, hence "direct access".
Collision Resolution: If there's a collision, it searches within the bucket, which is usually a small set.

Memory Trade-offs

Lists: Use memory efficiently but searching is slower.
Dictionaries: Use more memory (for buckets) but offer faster lookups.

Why Dictionaries Are Faster for Lookups

Reduced Search Space: Instead of searching the entire dataset, you only search one bucket.
Constant-Time Operations: Hash computation and bucket access are typically O(1) operations.
Load Factor: Dictionaries are often implemented to keep the number of elements per bucket low, maintaining efficiency.

Deep Dive: Hash Function's Role

Deterministic: The same key always produces the same hash.
Distribution: A good hash function distributes keys evenly across buckets.
Speed: Hash computation must be fast to maintain overall efficiency.

In the Context of Two Sum

List Approach: Would require checking each element for each number, leading to O(n^2) time complexity.
Dictionary Approach: Each lookup is O(1) on average, leading to O(n) overall as we only pass through the list once.

# Time Complexity Comparison

Nested Loop: 24 operations
Hash Table: 12 operations

The hash table approach is indeed faster, even in this small example. The difference becomes much more significant as the list size grows.
Scalability
For a list of size n:

Nested Loop: (n * (n-1) / 2) * 4 operations

For n = 1000, that's 1,998,000 operations


Hash Table: n * 3 operations

For n = 1000, that's 3,000 operations



This demonstrates why the hash table approach is considered O(n) while the nested loop is O(n^2).

```python
input_List = [2,7,11,15,4,3,10]
target = 5
empty_Dictionary = {}
result = []

for i,value in enumerate(input_List):
     if not (target - value) in empty_Dictionary.keys():
         empty_Dictionary[value] = i
     if (target - value) in empty_Dictionary.keys():
         result.append(empty_Dictionary[(target-value)])
         result.append(i)
         break
    
print(result)
```

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        empty_Dictionary = {}

        #length constraints
        if len(nums) < 2 or len(nums) > 10**4:  
            print("Length of the list is out of bounds")
            return [] 
        
        #target constraints
        if target < -10**9 or target > 10**9:
            print("Target is out of bounds")
            return []
        
        for index,value in enumerate(nums):

            #max value in list contraints
            if nums[index] < -10**9 or nums[index] > 10**9:
                print(f"The value of {nums[index]} is out of bounds")
                return []
                
            if not (target - value) in empty_Dictionary.keys():
                empty_Dictionary[value] = index
                continue

            if (target - value) in empty_Dictionary.keys():
                return [empty_Dictionary[(target-value)],index]
                
        print(f'No target value found within the list')
        return []
```
